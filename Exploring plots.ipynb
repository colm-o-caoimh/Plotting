```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use("ggplot")
```

```python
plt.plot([1, 2, 3], "r-", label="Red Line")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], "r-", label="Red Line")
ax.legend()
```
